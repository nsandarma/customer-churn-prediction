## Distribution

```python
def plot_dist(data,x,hue,type,title):
    if type == 'cat':
        ax = sns.countplot(x=data[x],
              hue=data[hue])
        for i in ax.containers:
            ax.bar_label(i,)
    else:
        ax = sns.boxplot(y= data[hue],hue=data[x])
    ax.set(title=title)
    return ax
```

### Categorical

```python
plot_dist(df_train,'churn','area_code','cat',title='area_code distribution');
print(df_train[["area_code","is_churn"]].groupby("area_code").mean().reset_index())
```

```python
plot_dist(df_train,'churn','international_plan','cat',title='international_plan distribution');
print(df_train[["international_plan","is_churn"]].groupby("international_plan").mean().reset_index())
```

```python
plot_dist(df_train,'churn','voice_mail_plan','cat',title='voice_mail_plan distribution');
print(df_train[["voice_mail_plan","is_churn"]].groupby("voice_mail_plan").mean().reset_index())
```
